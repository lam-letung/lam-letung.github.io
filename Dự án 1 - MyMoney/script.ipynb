# [dbo].[sp_ChiNam]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ChiNam]    Script Date: 04/12/2021 2:41:26 CH ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROC [dbo].[sp_ChiNam](@TenTK VARCHAR(50), @NamChi int)
AS BEGIN
	SELECT 
			lc.TenLC TenLoaiChi,
			SUM(kc.SoTienKC) TongChi,
			MIN(kc.SoTienKC) ThapNhat,
			MAX(kc.SoTienKC) CaoNhat,
			AVG(kc.SoTienKC) TrungBinh,
			MONTH (kc.NgayChi) ThangChi,
			YEAR(kc.NgayChi) NamChi
	FROM [dbo].[LoaiChi] lc
			JOIN  [dbo].[KhoanChi] kc ON lc.MaLC = kc.MaLC
			JOIN [dbo].[TaiKhoan] tk ON tk.TaiKhoan = kc.MaTK
	WHERE kc.MaTK = @TenTK AND YEAR(kc.NgayChi)= @NamChi
	GROUP BY lc.TenLC , kc.NgayChi
	ORDER BY ThangChi desc, NamChi desc
	

END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[sp_ThuNam]    Script Date: 04/12/2021 2:41:26 CH ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROC [dbo].[sp_ThuNam](@TenTK VARCHAR(50), @NamThu int)
AS BEGIN
	SELECT 
			lt.TenLT TenLoaiThu,
			SUM(kt.SoTienKT) TongThu,
			MIN(kt.SoTienKT) ThapNhat,
			MAX(kt.SoTienKT) CaoNhat,
			AVG(kt.SoTienKT) TrungBinh,
			MONTH (kt.NgayThu) ThangThu,
			YEAR(kt.NgayThu) NamThu
	FROM [dbo].[LoaiThu] lt 
			JOIN [dbo].[KhoanThu] kt ON lt.MaLT = kt.MaLT
			JOIN [dbo].[TaiKhoan] tk ON tk.TaiKhoan = kt.MaTK
	WHERE kt.MaTK = @TenTK  AND YEAR(kt.NgayThu)= @NamThu
	GROUP BY lt.TenLT , kt.NgayThu
	ORDER BY ThangThu desc, NamThu desc
	

END
GO
```
